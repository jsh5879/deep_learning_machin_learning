```
#odds.py
import numpy as np
import matplotlib.pyplot as plt

def odds(p) :  #시험
	return  p / (1 - p) 

p = np.arange(0.001, 0.999, 0.001)

z = [odds(n) for n in p]

plt.plot(p, z)
plt.ylim([0, 200.0])
plt.xlabel("p")
plt.ylabel("odds")

plt.show()
```

```
#logit.py
import numpy as np

import matplotlib.pyplot as plt

def logit1(p) :
	return  np.log( p / (1 - p)  )

#p = np.linspace(start=0.001, stop=0.999, num=1000)

p = np.arange(0.001, 0.999, 0.001)
z = [logit1(n) for n in p]

plt.grid()
plt.plot(p, z)
plt.xlabel("p")
plt.ylabel("logit")

plt.show()
```

```
#w가 다를때
import numpy as np 
import matplotlib.pyplot as plt 

def sigmoid(wx):
    return 1/(1+np.exp(-wx))

wx = np.arange(-6.0, 6.0, 0.1)
s1 = sigmoid(0.5*wx)
s2 = sigmoid(1.0*wx)
s3 = sigmoid(2.0*wx)

plt.grid()
plt.plot(wx, s1, 'r') # W = 0.5 
plt.plot(wx, s2, 'g') # W = 1 
plt.plot(wx, s3, 'b',  linestyle='--' ) # W = 2 
plt.title('Logistic | Sigmoid function')
plt.show()
```

```
#베르누이.py

import scipy as sp
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

# 코드로 확인하자
theta = 0.6 # 1이 나올 확률
rv = bernoulli(theta) # 확률변수 생성
type(rv)
rv.pmf([0,1])
xx = [0, 1]
plt.bar(xx, rv.pmf(xx), color='bg')
plt.xlim(-1, 2)
plt.ylim(0, 1)
plt.xticks([0, 1], ["x=0", "x=1"])
plt.ylabel("P(x)")
plt.title("Bernoulli distribution(PMF)")
plt.show()
```

```
#보스턴집값
#https://snowdeer.github.io/machine-learning/2018/01/09/predict-linear-values-using-machine-learning/
# data : https://snowdeer.github.io/assets/machine-learning/house_price.csv
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split  #sklearn 데이터관린

import pandas as pd

df = pd.read_csv('https://snowdeer.github.io/assets/machine-learning/house_price.csv', delim_whitespace=True, header=None)

data_set = df.values  #행, 열로 바꿔줌 [ [], [], [] ..... [] ]
X = data_set[:, 0:13]  #전체행, 0~12열
Y = data_set[:, 13]    #전체행, 13열

X_train, X_validation, Y_train, Y_validation = train_test_split(X, Y, test_size=0.2)    #실제데이터 >> 훈련데이터(0.8) + 유효성데이터(0.2) 로 나눔

model = Sequential()
model.add(Dense(100, input_dim=13, activation='relu'))  #13개 입력 100개로 퍼트림
model.add(Dense(6, activation='relu'))  #100개입력 6개출력
model.add(Dense(1))   #6개입력 1개출력

model.compile(loss='mean_squared_error', optimizer='adam')

model.fit(X_train, Y_train, epochs=200, batch_size=10)   #하이퍼파라미터 = 사용자입력

#predic = np.
#print(model.predict())

Y_prediction = model.predict(X_validation).flatten()

for i in range(10):
  real_price = Y_validation[i]
  predicted_price = Y_prediction[i]
  print('Real Price: {:.3f}, Predicted Price: {:.3f}'.format(real_price, predicted_price))
```

```
#영화긍정부정
#https://ssungkang.tistory.com/entry/%EB%94%A5%EB%9F%AC%EB%8B%9D05imdb-%EC%98%81%ED%99%94-%EB%A6%AC%EB%B7%B0-%EC%9D%B4%EC%A7%84-%EB%B6%84%EB%A5%98
#https://antilibrary.org/1965

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.layers import Dense

from keras.datasets import imdb
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words = 10000)

#print(train_data[0]) # 인덱스0 긍정적인 리뷰로 분류되어 있다.

word_index = imdb.get_word_index()
reverse_word_index = dict(
    [(value,key) for (key, value) in word_index.items()]
)
decode_review = ' '.join(
    [reverse_word_index.get(i-3, '?') for i in train_data[0]]
)

def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.
    
    print(results.shape)
    return results


x_train = vectorize_sequences(train_data)
x_test = vectorize_sequences(test_data)
print(x_train[:5, :])   #원핫인코딩, 문자열데이터를 숫자로 변환

y_train = np.asarray(train_labels).astype('float32')
y_test = np.asarray(test_labels).astype('float32')

model = Sequential()
model.add(Dense(16,activation='relu', input_shape=(10000,)))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))  #최종계층, softmax


model.compile(optimizer='rmsprop',
             loss='binary_crossentropy',
             metrics=['accuracy'])

x_val = x_train[:10000]   #검증(테스트)데이터 
partial_x_train = x_train[10000:]  #훈련데이터
y_val = y_train[:10000]
partial_y_train = y_train[10000:]

history = model.fit(partial_x_train,
                    partial_y_train,
                    epochs = 20,  #크면좋음
                    batch_size = 512, #작으면좋음
                    validation_data=(x_val, y_val))

history_dict = history.history
history_dict.keys()  #dict_keys(['val_loss', 'val_acc', 'loss', 'acc'])

import matplotlib.pyplot as plt
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1,len(loss) + 1)
plt.plot(epochs, loss, 'r', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title('Training Loss vs Validation Loss')
plt.show()   
```
